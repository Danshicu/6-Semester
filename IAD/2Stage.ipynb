```python
#https://www.kaggle.com/datasets/gpreda/covid19-tweets
dataframe = pd.read_csv('covid19_tweets.csv', sep=',')
dataframe.info()
```

```python
dataframe['date'] = pd.to_datetime(dataframe['date'])
dataframe
```

```python
actual_date = datetime.now() - timedelta(days = 1300)
actual_tweets = [row for index, row in dataframe.iterrows() if row['date'] >= actual_date]
tweets = pd.DataFrame(actual_tweets, columns = dataframe.columns)
tweets
```
